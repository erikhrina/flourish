# Model Conversion
Jupyter Notebook used to convert the develop Machine Learning model for Flutter application use.

### **Setup:**

Before running this notebook, make sure to have the final model saved at `/ModelDev/models/` as `plant_classifier.keras`.

---

```python
# Load the model
from keras.models import load_model

classifier = load_model('./models/plant_classifier.keras')
classifier.summary()
```

```python
# Save the model as .h5
classifier.save('./models/plant_classifier.h5')
```

```python
import tensorflow as tf

converter = tf.lite.TFLiteConverter.from_keras_model(classifier)
tflite_model = converter.convert()
open('./models/plant_classifier.tflite', 'wb').write(tflite_model)
```
